```python
# load in WikiHow dataset 
df_1024 = pd.read_csv('gdrive/MyDrive/SNLP_Coursework/WikiHow_sample_in1024.csv')
n = len(df_1024) / 2
print(n)
```

```python
# df_1024['train_length'] = 'empty'
# df_1024['summary'] = 'empty'
# df_1024['summary_length'] = 'empty'
# df_1024.head()
```

```python
def preprocess(text):
  preprocess_text = text.strip().replace("\n","")
  return len(preprocess_text.split())
```

```python
model = Summarizer()
```

```python
myText = df_1024['text'][:]
allText = []

for i, text in enumerate(myText):
  allText.append(text)

print(len(allText))
```

```python
allSummaries = []

for i, text in enumerate(allText):
  if i > 19 and i < n + 1:
    train_length = preprocess(text)
    m = round(0.33 * train_length)
    summary = ''.join(model(text, max_length=m))
    allSummaries.append(summary)
    df_1024['train_length'][i] = train_length
    df_1024['summary'][i] = summary
    df_1024['summary_length'][i] = len(summary.split())
    if i % 10 == 0:
      df_1024.to_csv('gdrive/MyDrive/SNLP_Coursework/WikiHow_sample_in1024_results_BERT.csv')
      print('SAVED!')
    print('Processed', i, 'out of', len(allText) / 2)
```
